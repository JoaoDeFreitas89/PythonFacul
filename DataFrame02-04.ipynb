#Data Frame
Estrutura para extração de dados a partir de uma fonte: Exemplo; CSV,XLSX, SQL..


```python
from re import X
import pandas as pd

df = pd.read_excel('/Idades.xlsx')
print(df)

df.plot.bar(x='Nome', y='Idade', color = 'red')
df.plot.line(x='Nome', y='Idade', color = 'blue')
df.plot.barh(x='Nome', y='Idade', color = 'green') #Plotagem dos graficos
df.plot.pie(y='Idade', labels=df['Nome'])
df.plot.pie(y='Idade', labels=df['Nome'], autopct='%1.1f%%')
df.plot.pie(y='Idade', labels=df['Nome'], autopct='%1.1f%%', legend=False)
df.plot.scatter(x='Nome', y='Idade', color = 'orange')
df.plot.box
df.plot.hist(bins=10)
```
